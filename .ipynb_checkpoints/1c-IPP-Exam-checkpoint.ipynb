```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define constants 
alpha = 1 / 137  # Fine-structure constant
s = 10
x = np.linspace(-0.99999,0.99999, 10000)  

# Define each term in the differential cross-section
def term1(x):
    return (1 + x**2)

def term2(x):
    return 2 * ((1 + x)**2+4) / (x - 1)**2

def term3(x):
    return 2 * (1 + x)**2 / (x - 1)

# Total differential cross-section
def total_cross_section(x):
    return (alpha**2 / (4 * s)) * (term1(x) + term2(x) + term3(x))

# Compute values for each component
term1_values = term1(x)
term2_values = term2(x)
term3_values = term3(x)
total_values = total_cross_section(x)

# Plotting
plt.figure(figsize=(15, 11))

# Plot each individual term
plt.plot(x, term1_values, label=r'$(1 + \cos^2\theta)$', linestyle='--', color='blue')
plt.plot(x, term2_values, label=r'$2((1 + \cos\theta)+4)/(\cos\theta - 1)^2$', linestyle='-.', color='green')
plt.plot(x, term3_values, label=r'$2(1 + \cos\theta)^2/(\cos\theta - 1)$', linestyle=':', color='red')

# Plot the total cross-section
plt.plot(x, total_values, label='Total Cross Section', color='black', linewidth=2)

plt.xlabel('cos(θ)', fontsize=14)
plt.ylabel('dσ/dΩ', fontsize=14)
plt.title('Angular Distribution of Bhabha Scattering Cross Section', fontsize=16)
plt.legend()

# Set axis limits
plt.xlim(-1, 1.1)  
plt.ylim(-20, 20) 
plt.grid(True)
plt.show()
```

The total cross-section diverges as cos(θ) approaches 1, due to the t-channel contribution, which leads to an enhancement of forward scattering. The s-channel provides a smooth, symmetric distribution, and its interference with the t-channel shapes the overall angular distribution.
